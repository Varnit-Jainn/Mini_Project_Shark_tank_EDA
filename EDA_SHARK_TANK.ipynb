```python
#Q10. How many sharks participated in this show and What were their names?
```

```python
col=df.columns.values
```

```python
a=0
print("Names of sharks participated are as follows : ")
for i in col:
    if "_present" in  i:
        x=i.split("_")
        print(x[0])
        a=a+1
print(f"number of sharks participated={a}")
```
